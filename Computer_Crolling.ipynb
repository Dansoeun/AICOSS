```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://www.knu.ac.kr/wbbs/wbbs/user/yearSchedule/index.action?menu_idx=43',headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
# 이 이후로 필요한 부분 추출

```

```python
# Finding the calendar section
calendar = soup.find('div', id='calendar')

# Extracting the month and year
month_year = calendar.find('dt').get_text(strip=True)
print(f"Month and Year: {month_year}")

# Extracting the list of events
events = calendar.find_all('li')

# Printing each event with the corresponding day
for event in events:
    day = event.find('span', class_='day').get_text(strip=True)
    description = event.get_text(strip=True).replace(day, '').strip()
    print(f"{day}: {description}")
```

```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
page = {}

for i in range(1, 11):
    url = f"https://cse.knu.ac.kr/bbs/board.php?bo_table=sub5_1&page={i}"
    response = requests.get(url, headers=headers)  # requests.get()을 사용하여 웹페이지 데이터를 가져옴

    soup = BeautifulSoup(response.text, 'html.parser')  # 가져온 데이터를 BeautifulSoup으로 파싱합

    notice = soup.find_all("div", {"class": "bo_tit"})  # 모든 bo_tit 클래스를 가진 div 태그 찾기

    # 제목만 추출
    titles = [n.get_text(strip=True) for n in notice]

    page.update({url: titles})  # 페이지 번호에 따른 제목들을 저장

print(page)
```
