```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}

data = requests.get('https://cse.knu.ac.kr/sub5_5.php',headers=headers)
soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

```python
calendarLink = ""

calendar_temp = soup.find_all("div",{"class":"lab_top"})
for element in calendar_temp :
    calendarLink = element.find_all('iframe')[0]['src']
    print(calendarLink)
```
