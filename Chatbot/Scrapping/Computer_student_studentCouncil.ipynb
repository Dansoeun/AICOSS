```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}

data = requests.get('https://cse.knu.ac.kr/bbs/board.php?bo_table=sub6_1_a&lang=kor',headers=headers)
soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

```python
student_council_content = soup.find_all("div",{"class":"basic_tbl_head tbl_wrap"})
print(student_council_content)
```

```python
print("[학생회 공지사항 게시글 목록]")

for element in student_council_content :
    council_content = element.find_all("a")
    for e in council_content:
        etext = e.get_text()
        etext = etext.replace("                                                                        ","")
        etext = etext.replace("\n","")
        print(etext)
```
