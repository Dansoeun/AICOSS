```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://cse.knu.ac.kr/bbs/board.php?bo_table=sub3_6&lang=kor',headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

```python
admission_temp = soup.find_all("div",{"class":"s_wrap"})
```

```python
for element in admission_temp :
    ad_main = element.find_all("tr")
    for tr in ad_main:
        tr_th = tr.find_all("th")
        for h in tr_th :
            print()
            print(h.get_text(),end = "")
        tr_td = tr.find_all("td")
        for d in tr_td :
            dtext= d.get_text()
            if dtext.isdigit() == True or dtext == "-" :
                print(dtext,end=" ")
            else :
                print()
                print(dtext,end=" ")
```

```python
faq_List = soup.find_all("ul",{"class":"faqList"})
print(faq_List)
```

```python
question_list = []
answer_list = []

def blank(e):
    if e == ".":
        print()
        print()
    elif len(e)>1:
        if e[-1] == "." :
            print()
            print()


for element in faq_List:
    
    faqQuestion = element.find_all("div",{"class":"faqQ"})
    faqAnswer = element.find_all("div",{"class":"faqA"})
    for felement in faqQuestion :
        faqQ = felement.find_all("div",{"class":"content"})
        for f in faqQ:
            print(f.get_text())
            question_list.append(f.get_text())
            
    for aelement in faqAnswer :
        faqA = aelement.find_all("span")
        for a in faqA:
            answer_list.append(a.get_text())
            print(a.get_text(),end="")
            blank(a.get_text())
```
