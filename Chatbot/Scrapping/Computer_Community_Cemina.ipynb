```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}

data = requests.get('https://cse.knu.ac.kr/bbs/board.php?bo_table=sub5_4&lang=kor',headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

```python
cemina_container_temp = soup.find_all("div",{"id":"sub_container"})
for element in cemina_container_temp:
    cemina_list = element.find_all("a")
    for c in cemina_list:
        ctext = c.get_text()
        ctext = ctext.replace("                                                                        ","")
        ctext = ctext.replace("\n","")
        ctext = ctext.replace("다음","").replace("맨끝","").replace("조회수","").replace("날짜","")
        
        for i in range(2,11):
            text1 = str(i)+"페이지"
            ctext = ctext.replace(text1,"")
        #ctext = ctext
        if ctext.replace(" ","") == None:
            continue
        print(ctext)
```
