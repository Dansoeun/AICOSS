```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}

data = requests.get('https://cse.knu.ac.kr/sub6_3.php',headers=headers)
soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

```python
club_temp = soup.find_all("div",{"id":"sub_container"})
for element in club_temp:
    etext = element.get_text().replace("\n\n","")
    print(etext)
```
