```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://cse.knu.ac.kr/sub3_2_a.php',headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
# 이 이후로 필요한 부분 추출
```

```python
#abeek이란
what_abeek = soup.find_all("div",{"class":"lab_top"})
print(what_abeek)
```

```python
for element in what_abeek :
    abeek = element.find("p")
    for p in abeek:
        print(p.get_text(),end="")
```

```python
#abeek교육 목표
abeek_g = soup.find_all("div",{"class":"data_top"})

for element in abeek_g:
    abeek_goal = element.find_all("p")
    for p in abeek_goal:
        print(p.get_text())
```

```python
#second div table : graduation credit
grad_credit = soup.find_all("div",{"class":"table_wrap_view"})
print(grad_credit)
```

```python
for element in grad_credit:
    grad_credit_th = element.find_all("th")
    grad_credit_td = element.find_all("td")
    for e in grad_credit_th:
        print(e.get_text())
    for f in grad_credit_td:
        print(f.get_text())
```

```python
#platform 2
#https://cse.knu.ac.kr/sub3_2_b.php
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://cse.knu.ac.kr/sub3_2_b.php',headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

```python
subjects_pf = soup.find_all("div",{"class":"table_wrap_view"})
print(subjects_pf)
```

```python
for element in subjects_pf:
    grade1 = element.find_all("th",{"rowspan":"3"})
    subject_td = element.find_all("td")
    for g in grade1 :
        print(g.get_text())
    
    for s in subject_td:
        print(s.get_text())
        
```
