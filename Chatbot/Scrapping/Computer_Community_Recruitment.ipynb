```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}

data = requests.get('https://cse.knu.ac.kr/bbs/board.php?bo_table=sub5_3_a&lang=kor',headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

```python
recruitment = soup.find_all("div",{"class":"basic_tbl_head tbl_wrap"})
print(recruitment)
```

```python
count = 0
for element in recruitment:
    recr_a = element.find_all("a")
    for h in recr_a:
        count+=1
        htext = h.get_text()
        htext = htext.replace("                                                                        ","")
        htext = htext.replace("\n","")
        print(htext)
        
```
