```python
#page2
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://cse.knu.ac.kr/sub3_4_b.php',headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

#print(soup)

subject_gsw2_temp = soup.find_all("div",{"class":"table_wrap_view"})
print(subject_gsw2_temp)
```

```python
subject_name = []
subtitles = []
ititles = []

for element in subject_gsw2_temp:
    subject_gsw2_td = element.find_all("td")
    subject_gsw2_th = element.find_all("th")
    subject_gsw2_i = element.find_all("i")
    for h in subject_gsw2_th:
        #print(h.get_text())
        subject_name.append(h.get_text())
    for d in subject_gsw2_td:
        #print(d.get_text())
        subtitles.append(d.get_text())
    for i in subject_gsw2_i:
        #print(i.get_text())
        ititles.append(i.get_text())

print(subject_name)
print(subtitles)
print(ititles)
```
